**Prompt:** A Python program to create an SQLite database, create an Employees table, insert employee records, and display the stored data using the Pandas DataFrame.

**Prompt:** to display the total number of employees in the Employees table using the SQL COUNT() function.

```python
# Total Employees
print("Total Employees")
display(pd.read_sql("""
SELECT COUNT(*) AS TotalEmployees
FROM Employees
""", conn))
```

**Prompt:** to calculate and display the average total working years of employees using the SQL AVG() function.

```python
print("Average Experience")
display(pd.read_sql("""
SELECT ROUND(AVG(Total_working_years),2) AS AverageExperience
FROM Employees
""", conn))
```

**Prompt:** to display the number of employees in each department using the SQL COUNT() function and GROUP BY clause.

```python
print("Department-wise Employee Count")
display(pd.read_sql("""
SELECT Department,
       COUNT(*) AS EmployeeCount
FROM Employees
GROUP BY Department
""", conn))
```

**Prompt:** to display the details of employees with a Performance Rating of 5 using the SQL WHERE clause.

```python
print("Display Employees with Performance Rating is 5")
display(pd.read_sql("""
SELECT *
FROM Employees
WHERE PerformanceRating = 5
""",conn))
```
